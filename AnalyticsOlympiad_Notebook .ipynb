```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
train.head()
```

```python
train.describe()
```

```python
train.dtypes
```

```python
y = train.OUTCOME
```

```python
data = pd.concat((train,test)).reset_index(drop = True).copy()
data = data.drop(columns=["ID","OUTCOME"],axis = 1)
data
```

```python
#Check whether there are any missing value in the table
missing_props = data.isna().mean(axis=0)
missing_props
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

cat_pipeline = Pipeline(
    steps=[
        ("ohe", OneHotEncoder(handle_unknown="ignore", sparse=False))
    ]
)

num_pipeline = Pipeline(
    steps=[
        ("scale", StandardScaler())
    ]
)

cat_cols = data.select_dtypes(exclude="number").columns
num_cols = data.select_dtypes(include="number").columns
```

```python
from sklearn.compose import ColumnTransformer

full_processor = ColumnTransformer(
    transformers=[
        ("numeric", num_pipeline, num_cols),
        ("categorical", cat_pipeline, cat_cols)
    ]
)
```

```python
data_processed = full_processor.fit_transform(data)
```

```python
train = data_processed[:len(train)]
test = data_processed[len(train):]
```

```python
import xgboost as xgb

xgb_cl = xgb.XGBClassifier()
print(xgb_cl)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_val,y_train,y_val = train_test_split(
    train, y, random_state = 42
)
```

```python
param_grid = {
    "max_depth": [3,4,5,7],
    "learning_rate": [0.1, 0.01, 0.05],
    "gamma": [0, 0.25, 1],
    "reg_lambda": [0, 1, 10],
    "scale_pos_weight": [1,3,5],
    "subsample": [0.8],
    "colsample_bytree": [0.5],
}
```

```python
from sklearn.model_selection import GridSearchCV

xgb_cl = xgb.XGBClassifier(objective="binary:logistic")

grid_cv = GridSearchCV(xgb_cl, param_grid, n_jobs=-1, cv=2, scoring="neg_log_loss")

_ = grid_cv.fit(X_train, y_train)
```

```python
grid_cv.best_score_
```

```python
grid_cv.best_params_
```

```python
xgb_cl = xgb.XGBClassifier(learning_rate=0.05, max_depth=3, reg_lambda=10, scale_pos_weight=1, subsample=0.8, gama=0, colsample_bytree=0.5)
```

```python
from sklearn.metrics import log_loss

xgb_cl.fit(X_train, y_train)

preds = xgb_cl.predict_proba(X_val)

print(log_loss(y_val, preds))
```

```python
xgb_cl.fit(train, y)
final_preds = xgb_cl.predict_proba(test)
```

```python
final_preds = final_preds[:,1]
```
